```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

## Data Preprocessing
### 1. Read in raw data
- train.csv.gz -> df_train_0
- test.csv.gz -> df_test_0

```python
###############################################
#            Data Preprocessing               #
###############################################

## 1. Read in raw data
# use panda to read in data 
df_train_0 = pd.read_csv("../train.csv.gz")
df_test_0 = pd.read_csv("../test.csv.gz")
```

```python
# view the train data
#df_train_0.head()
```

```python
# view the test data
#df_test_0.head()
```

### 2. clean data

```python
## 2. Clean data
# 2.1 combine train and test for easier feature engineering
# (1) delete 'Id' column
df_test = df_test_0.drop(['Id'], axis=1)
# (2) delete 'gap' column
df_train = df_train_0.drop(['gap'], axis=1)
# (3) combine train and test
df_all = pd.concat((df_train, df_test), axis=0)

# 2.2 Drop the 'smiles' column
df_all = df_all.drop(['smiles'], axis=1)
```

```python
# view the cleaned data
#df_all.head()
```

### 3. specify x and y

```python
# Specify x and y
# 3.1 specify gap values as y
Y_train = df_train_0.gap.values

# 3.2 specify x values
# index of row where testing examples start
test_idx = df_train.shape[0]
# split x_train and x_test
X_train = df_all.values[:test_idx]
X_test = df_all.values[test_idx:]

# 3.3 view data structure
print("Train features:", X_train.shape)
print("Train gap:", Y_train.shape)
print("Test features:", X_test.shape)
```

## Feature Selection
Use baseline model to select impotant features
- Linear Regression: drop coef=0
- Random Forest: drop importance=0

### 1. Fit base line models

```python
###############################################
#             Feature Selection               #
###############################################
# Use baseline model to select impotant features
# 1. Fit base line models
# 1.1 Fit Linear Regressoin
LR = LinearRegression()
LR.fit(X_train, Y_train)
#LR_pred = LR.predict(X_test)

# 1.2 Fit Random Forest
RF = RandomForestRegressor()
RF.fit(X_train, Y_train)
#RF_pred = RF.predict(X_test)

# 1.3 Calculate initial R^2
print("R^2 for Linear Regression is: ",LR.score(X_train, Y_train))
print("R^2 for Random Forest is: ",RF.score(X_train, Y_train))
```

### 2. Drop vaccant predictors

```python
# 2. Drop predictors that are unimportant in both LR and RF
LR_coef = LR.coef_
RF_imp = RF.feature_importances_
i0 = (LR_coef == 0) & (RF_imp == 0)
X_train1 = X_train[:,i0 == False]
X_test1 = X_test[:,i0 == False]
# view data structure after drop
print("Train features after 1 drop:", X_train1.shape)
print("Train gap:", Y_train.shape)
print("Test features after 1 drop:", X_test1.shape)
```

### 3. Refit baseline models

```python
# 3. Refit base line models
# 3.1 Refit Linear Regressoin
LR1 = LinearRegression()
LR1.fit(X_train1, Y_train)

# 3.2 Refit Random Forest
RF1 = RandomForestRegressor()
RF1.fit(X_train1, Y_train)

# 3.3 Recalculate R^2
#print(LR1.score(X_train1, Y_train))
#print(RF1.score(X_train1, Y_train))
print("R^2 for refitted Linear Regression is: ",LR1.score(X_train1, Y_train))
print("R^2 for refitted Random Forest is: ",RF1.score(X_train1, Y_train))
```

### 4. Drop features < 0

```python
# 4.1 Dop features with RF importance < 0
RF_imp1 = RF1.feature_importances_
LR_coef1 = LR.coef_
i1 = (RF_imp1 > 0)
X_train2 = X_train1[:,i1]
X_test2 = X_test1[:,i1]
```

```python
# 4.2 view data structure
print("Train features after 2 drops:", X_train2.shape)
print("Train gap:", Y_train.shape)
print("Test features after 2 drops:", X_test2.shape)
```

## Generate Prediciton on Test

```python
# function to generate prediciton on test set
def write_to_file(filename, predictions):
    with open(filename, "w") as f:
        f.write("Id,Prediction\n")
        for i,p in enumerate(predictions):
            f.write(str(i+1) + "," + str(p) + "\n")
```
